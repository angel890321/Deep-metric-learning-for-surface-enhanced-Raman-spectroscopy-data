## ASTIII

```python
import os
import pandas as pd
import random
from traceback import print_exc
from preprocessing import remove_baseline


# 初始化一個空的 DataFrame
df_all_values_astiii = pd.DataFrame()

# 初始化類別的列表
classes = []
patient_ids = []
row_index = 0  # 初始化行的編號
# 設定篩選的範圍和間隔
start_wavenumber = 404
end_wavenumber = 2000
target_wavenumbers = np.arange(start_wavenumber, end_wavenumber+1, 4)

# 遍歷每個類別資料夾
result = "\n"

# 遍歷每個第二層類別資料夾
for category in os.listdir(root_dir_2):
    category_path = os.path.join(root_dir_2, category)
    if os.path.isdir(category_path):
        # 遍歷第三層資料夾
        for subcategory in os.listdir(category_path):
            subcategory_path = os.path.join(category_path, subcategory)
            if os.path.isdir(subcategory_path):

                for file_name in os.listdir(subcategory_path):
                    file_path = os.path.join(subcategory_path, file_name)
                    if os.path.isfile(file_path):
                        
                        try:
                            debaselined,single_spec= remove_baseline(file_path, 50, 0.5)


                            df = pd.DataFrame(debaselined, columns=['wavenumber', 'value'])
                            
                            if single_spec == True:
                                # 篩選 wavenumber 介於 400 到 2000 的範圍
                                filtered_data = df[(df['wavenumber'] >= start_wavenumber) & (df['wavenumber'] <= end_wavenumber)]

                                target_values = []

                                # 對每個目標波數，找到最接近的實際數據點或填入 0
                                for target in target_wavenumbers:

                                    filter_value = filtered_data[filtered_data['wavenumber'] == float(target)]['value'].values
                                    
                                    target_values.append(filter_value[0])

                                # 創建一個 DataFrame 以 row_index 為行名稱
                                df_single_row = pd.DataFrame([target_values], index=[row_index])

                                # 將該行添加到 df_all_values_astiii
                                df_all_values_astiii = pd.concat([df_all_values_astiii, df_single_row])

                                # 將類別名稱加入 classes 列表
                                classes.append(category)
                                patient_ids.append(subcategory)
                                row_index += 1  # 增加行的編號
                            else:
                                print("->無法處理")
                                result += "無法處理: " + file_path + "\n"
                                print_exc()
                            
                        except Exception as e:
                                print(f"Error reading file {file_path}: {e}")



```

```python
# 將 Classes 資訊添加到 DataFrame 作為新的一列
df_all_values_astiii['Classes'] = classes
df_all_values_astiii['Patient_ID'] = patient_ids

# 將 'Classes' 列移動到 DataFrame 的第一列
df_all_values_astiii = df_all_values_astiii[['Classes', 'Patient_ID'] + [col for col in df_all_values_astiii.columns if col not in ['Classes', 'Patient_ID']]]
# 檢查結果
print(df_all_values_astiii.head())
```

```python

# 呼叫函數移除 NaN 值
df_astiii_cleaned = df_all_values_astiii.dropna()
```
